```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
import statsmodels.api as sm
import itertools
import warnings
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from mpl_toolkits import mplot3d
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from IPython.display import clear_output
import matplotlib.cm as cm
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as shc
```

```python
data=pd.read_csv('Mall_Customers.csv')
```

```python
data.head()
```

```python
sns.pairplot(data=data[data.columns[1:]],hue='Gender')
plt.show()
```

By seeing at this plot we can discad the fact that being a female or male customer affects in our data

```python
x=data[data.columns[2:]]
X=StandardScaler().fit_transform(x)
```

```python
pca = PCA(n_components=0.95)
principalComponents = pca.fit_transform(X)
```

```python
for i in range(0,180,4):
    fig = plt.figure(figsize=(12,7))
    ax = plt.axes(projection='3d')
    ax.scatter3D(principalComponents[:,0],principalComponents[:,1],principalComponents[:,2],c=pd.get_dummies(data, columns=['Gender'])['Gender_Female'].values,cmap='coolwarm')
    ax.view_init(35, i)
    plt.show()
    clear_output(wait=True)
```

```python
plt.scatter(principalComponents[:,0],principalComponents[:,1],c=pd.get_dummies(data, columns=['Gender'])['Gender_Female'].values)
```

```python
range_n_clusters = np.arange(2,30)
for n_clusters in range_n_clusters:
    fig, (ax1) = plt.subplots(1, 1,figsize=(20,5))
    #fig.set_size_inches(15, 10)
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)
    
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)
    sample_silhouette_values = silhouette_samples(X, cluster_labels)
    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("K = {}".format(n_clusters))
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])


plt.show()
```

```python
range_n_clusters = list(np.arange(2,30))
print(range_n_clusters)
savergs=[]
for n_clusters in range_n_clusters:

    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    
    silhouette_avg = silhouette_score(X, cluster_labels)
    savergs.append(silhouette_avg)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)

plt.plot(range_n_clusters,savergs)
plt.xlabel("K")
plt.ylabel("Average silhouette score")
```
